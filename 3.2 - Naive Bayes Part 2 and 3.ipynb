```python
import os
import io
import numpy 
from pandas import DataFrame
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import pyearth
```

```python
def readFiles(path):
    for root,dirnames,filenames in os.walk(path):
        for filename in filenames:
            path=os.path.join(root,filename)
            
            inBody=False
            lines =[]
            f = io.open(path,'r',encoding='latin1')
            for line in f:
                if inBody:
                    lines.append(line)
                elif line=='\n':
                    inBody=True
            f.close()
            message = '\n'.join(lines)
            yield path,message    
```

```python
def dataFrameFromDirectory(path,classification):
    rows =[]
    index=[]
    for filename,message in readFiles(path):
        rows.append({'message': message,'class': classification})
        index.append(filename)
        
    return DataFrame(rows,index=index)
```

## Cross-validation

```python
from random import randrange
 
       
# Split a dataset into k folds
def cross_validation_split(dataset, folds=3):
    dataset_split = list()
    dataset_copy = list(dataset)
    fold_size = int(len(dataset) / folds)
    for i in range(folds):
        fold = list()
        while len(fold) < fold_size:
            index = randrange(len(dataset_copy))
            fold.append(dataset_copy.pop(index))
        dataset_split.append(fold)
    return dataset_split
```

### Create concatenated dataset (xtrain + ytrain)

```python
xdata=x_train.toarray()
print('xdata:')
#print(xdata)
print(xdata.shape)
print('\n')

y_data=y_train
print('y_data:')
#print(y_data)
print(y_data.shape)
print('\n')


alldata=np.c_[xdata,y_data]
print("ALL DATA\n")    
#print(alldata)
print(alldata.shape)
```

### Choose value of k for k-fold cross validation

```python
number_folds=10  #choose value of k

number_rows=xdata.shape[0]
fold_size = round(number_rows / number_folds)
print('Fold size:',fold_size)
```

### Training and testing on different folds

```python
folds = cross_validation_split(alldata, folds=number_folds)

accuracies=[]

for i in range(number_folds):
     
    #------TEST data------
    print('\n Testing on fold',i)
    testfold=np.array(folds[i])
    #print('Test fold:\n',testfold)
    
    
    
    #------TRAIN data------
    #remove test fold index
    trainindices=np.arange(number_folds).tolist()
    trainindices.pop(i)
    print('Training on folds',trainindices)
    
    #print("Train dataset:")
    train_data=np.zeros((0,alldata.shape[1])) 
    for j in trainindices:
        train_data=np.vstack((train_data,folds[j]))
    #print(train_data)
    
    
    
    
    #train and test data is ready but its the x and y data concatenated.we must split them  
    final_x_train=train_data[:,0:(train_data.shape[1]-1)]
    final_y_train=train_data[:,(train_data.shape[1]-1)]
    
    
    
    
    #train Naive Bayes model
    classifier = MultinomialNB()
    classifier.fit(final_x_train,final_y_train)

    predictions=classifier.predict(x_test)

    naive_bayes_accuracy_score=accuracy_score(y_test,predictions)

    print('ACCURACY:',naive_bayes_accuracy_score)
    
    accuracies.append(naive_bayes_accuracy_score)
    
```

```python
average_accuracy=np.average(accuracies)
average_accuracy
```

The cross-validation results seem pretty **reliable** ,compared to evaluating accuracy on a separate test set.  
This however, is an extreme case where the accuracy is extremely good and it would be interesting to see what are the deviations from the test set accuracy, when the model is not so good.  
Still though, cross-validation is a very popular and widely used model validation technique.
